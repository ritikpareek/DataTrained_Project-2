# Regression Process

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
def max_r2_score(rgr,x,y):
    max_score=0
    for r in range(42,101):
        
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=r)
        rgr.fit(x_train,y_train)
        pred=rgr.predict(x_test)
        score=r2_score(y_test,pred)
        print('score at',r,'is',score)
        if score>max_score:
            max_score=score
            r_state=r
    print('max score at the r state of',r_state,'is',max_score)
            
    return r_state
        
```

```python
lr=LinearRegression()
```

```python
sc=max_r2_score(lr,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=64)
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
print(r2_score(y_test,pred))
```

```python
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

# Lasso

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lsr=Lasso()
pr={'alpha':[.001,.01,.1,1]}
gs=GridSearchCV(lsr,pr,cv=10)
gs.fit(x,y)
gs.best_params_
```

```python
lsr=Lasso(alpha=.001)
sc=max_r2_score(lsr,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=64)
lsr.fit(x_train,y_train)
pred=lsr.predict(x_test)
```

```python
print(r2_score(y_test,pred))
```

```python
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

# Ridge

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
rdr=Ridge()
pr={'alpha':[.001,.01,.1,1]}
gs=GridSearchCV(rdr,pr,cv=10)
gs.fit(x,y)
gs.best_params_
```

```python
rdr=Ridge(alpha=1)
sc=max_r2_score(rdr,x,y)
```

```python
print(r2_score(y_test,pred))
```

```python
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```
