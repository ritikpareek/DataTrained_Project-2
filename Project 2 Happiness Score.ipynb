# Removing Outliers

```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(df))
```

```python
df_new=df[(z<3).all(axis=1)]
```

# Removing skewness 

```python
df_new.skew()
```

for making a good model the skewness should be less than .055 but in this dataset some variables are not normally distributed we have to make them normal

```python
for col in df_new.columns:
    if df_new.skew().loc[col]>0.55:
        df_new[col]=np.log1p(df_new[col])
```

# setting the x and y variable

```python
x=df_new.drop('Happiness Score',axis=1)
```

```python
x.shape
```

```python
y=df_new['Happiness Score']
```

```python
y=y.values.reshape(y.shape[0],1)
```

# standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
x=scale.fit_transform(x)
```

```python
x.shape
```

# Regression Process

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
def max_r2_score(rgr,x,y):
    max_score=0
    for r in range(42,101):
        
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=r)
        rgr.fit(x_train,y_train)
        pred=rgr.predict(x_test)
        r_sc=r2_score(y_test,pred)
        if r_sc>max_score:
            max_score=r_sc
            r_state=r
            
    print('max score at random_state of',r_state,'is',max_score)
    return r_state
```

```python
lr=LinearRegression()
```

```python
sc=max_r2_score(lr,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=68)
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lsr=Lasso()
pr={'alpha':[.001,.01,.1,1]}
gs=GridSearchCV(lsr,pr,cv=10)
gs.fit(x,y)
gs.best_params_
```

```python
lsr=Lasso(alpha=.001)
sc=max_r2_score(lsr,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=68)
lsr.fit(x_train,y_train)
pred=lsr.predict(x_test)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

# KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor()
pr={'n_neighbors':range(1,30)}
gs=GridSearchCV(knn,pr,cv=5)
gs.fit(x,y)
gs.best_params_
```

```python
knn=KNeighborsRegressor(n_neighbors=7)
sc=max_r2_score(knn,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=65)
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

# GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gb=GradientBoostingRegressor()
pr={'learning_rate':[.001,.01,.1,1],'n_estimators':[10,100,500,1000]}
gs=GridSearchCV(gb,pr,cv=5)
gs.fit(x,y)
gs.best_params_
```

```python
gb=GradientBoostingRegressor(learning_rate=0.1,n_estimators=1000)
sc=max_r2_score(gb,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=93)
gb.fit(x_train,y_train)
pred=gb.predict(x_test)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

# AdaBoostRegressor
