# Crime Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv('crime.csv')
```

```python
df=pd.DataFrame(ds)
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df['Area_Name'].unique()
```

```python
df['unreported_cases']=df['Victims_of_Rape_Total']-df['Rape_Cases_Reported']
```

```python
df.head()
```

```python
u=df.groupby(['Area_Name']).sum()
u.drop('Year',axis=1,inplace=True)
u['unreported_cases'].max()
```

```python
u.sort_values(by='unreported_cases',ascending=False)
```

Highest unreported cases are found in Maharastra

```python
u.sort_values(by='Victims_of_Rape_Total',ascending=False)
```

we can see that highest rape case are found in MP and lowest rape case are found in Lakshadweep

```python
u.sort_values(by='Victims_Above_50_Yrs',ascending=False)
```

victims of above 50 age are highest in MP and lowest in A&N

```python
u.sort_values(by='Victims_Upto_10_Yrs',ascending=False)
```

Highest child rape are found in Maharastra
